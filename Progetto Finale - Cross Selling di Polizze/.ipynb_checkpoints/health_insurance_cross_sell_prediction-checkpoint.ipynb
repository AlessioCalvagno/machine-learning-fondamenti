### Feature distributions
Here some charts are used to have a rough idea about features distribution, if there is imbalance, or some other information useful for prediction problem.

For qualitative features the relative frequency distribution in shown, for quantitative ones, some statistical properties (mean, standard deviation and quartiles) are provided along with a plot of feature's distribution.
If the feature is discrete the distribution is shown as an histogram + boxplot, else a violinplot is provided.

Note: 
For Region_Code and Policy_Sales_Channel features, we use seaborn histplot, because these variables have a lot of categories, and with the custom function _plot_frequency_distribution_ the plot is too messy.

```python
#define a custom function to make relative frequency distributions for 
#qualitative features. In charts, bins are frequency sorted

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

#some graphics settings
plt.rcParams["figure.figsize"] = (16,10)
#plt.rcParams["font.size"] = 50
sns.set_theme(style='dark')

def plot_frequency_distribution(X, feature_name=None):
    """
    Make a bar chart of relative frequency distribution for a qualitative
    feature.
    The relative frequency values are printed in console as percentages.
    Prameters:
    - X : array-like or pandas.Series object
        Feature values.
    - feature_name: str or None
        Feature name, used for figure title.
        If None, no title is displayed. 
        Default None.
    """
    FONT_SIZE = 16
    
    relative_frequency = X.value_counts(normalize = True)*100
    #relative_frequency.sort_index(inplace=True)
    print(relative_frequency)
    ax = relative_frequency.plot.bar(rot=45, fontsize=FONT_SIZE)
    pf = PercentFormatter(xmax=100,decimals=2)
    ax.yaxis.set_major_formatter(pf)
    ax.bar_label(ax.containers[0], fmt='%.2f%%', fontsize=FONT_SIZE)
    plt.ylabel('Frequency (%)', fontsize=FONT_SIZE)
    plt.xlabel('Categories',fontsize=FONT_SIZE)
    plt.ylim((0,100))
    if feature_name != None:
        plt.title(f"Relative frequency distribution for {feature_name} feature",
                 fontsize=FONT_SIZE)
    plt.show()
    
 
```

```python
def plot_quantitative_distribution(X, feature_name=None, discrete=False):
    """
    Make plot for distribution of a quantiative feature.
    If the feature is discrete, a pair of histogram and boxplot is shown,
    otherwise a violinplot is shown.
     Prameters:
    - X : array-like or pandas.Series object
        Feature values.
    - feature_name: str or None
        Feature name, used for figure title.
        If None, no title is displayed. 
        Default None.
    - discrete: bool
        If true, X is treated as a discrete feature (histogram + boxplot),
        else X is treared as continous (violinplot).
        Default False
    """
    FONT_SIZE = 16  
    if(discrete):
        #histogram
        fig, axs = plt.subplots(2,1,sharex=True)
        sns.histplot(x=X, discrete=True, ax = axs[0])
        axs[0].set_ylabel("Count", fontsize=FONT_SIZE)
        axs[0].tick_params(axis='y', labelsize=FONT_SIZE)
        
        #boxplot
        sns.boxplot(x=X,width=.5,ax = axs[1])
        plt.tight_layout()
    else:
        sns.violinplot(x=X)
        sns.despine(left=True)
        
    plt.xticks(fontsize=FONT_SIZE)
    if feature_name != None:
        plt.xlabel(feature_name, fontsize=FONT_SIZE)
        plt.suptitle(f"{feature_name} distribution", fontsize=FONT_SIZE)
    plt.show()
```

## Feature selection

Another step useful for feature selection is computing the correlation matrix and see the relationship between features and taget.
Since we want to use a logistic regression classifier, features should not be too much each other correlated (see [multicollinearity](https://www.wikiwand.com/en/Multicollinearity)) and there should be high correlation between feature and target.

```python
#some data preprocessing before going further
data_clean = data.drop("Driving_License",axis=1)
gender_mapping = {"Male":0, "Female":1}
data_clean["Gender"] = data_clean["Gender"].map(gender_mapping)

veichle_age_mapping = {"< 1 Year":0, "1-2 Year":1, "> 2 Years":2}
data_clean["Vehicle_Age"] = data_clean["Vehicle_Age"].map(veichle_age_mapping)

vehicle_damage_mapping = {"No":0, "Yes":1}
data_clean["Vehicle_Damage"] = data_clean["Vehicle_Damage"].map(vehicle_damage_mapping)
```

For each variable, it's a good practice do a multivariate analysis, in order to establish if it's suitable for prediction. 
As visual metric we plot the conditional distribution through an histogram. The best case is when the histograms of the two distributions are well separated. 
As numeric metric, we perform chi-squared test, to check if there is an association between target and features.


```python
#custom function to plot multivariate histogram

def plot_multi_hist(data,feature_name,target_name):
    """
    Plot a multivariate histogram, in seaborn style. 
    For each class in target, an histogram is computed and displayed
    with an unique color. 
    This plot is ufesul to determine if a feature is suitable for target 
    prediction.
    
    Parameters:
    - data: pandas.DataFrame object.
        The dataset containing feature and target.
    - feature_name: str.
        Name of the feature to plot.
    - target_name: str.
        Name of target feature, as in data object. 
    """
    
    FONT_SIZE = 16
    sns.histplot(data = data, x = feature_name,
                 hue = target_name, multiple="dodge")
    plt.yticks(fontsize = FONT_SIZE)
    plt.xticks(fontsize = FONT_SIZE)
    plt.xlabel(feature_name, fontsize = FONT_SIZE)
    plt.ylabel("Count", fontsize = FONT_SIZE)
    plt.title(f"Conditional histogram for {feature_name}")
    plt.show()
    
```

```python
from scipy.stats import chi2_contingency

def chi_test_feature_target(feature, target, categorical=False, nbins=10):
    """
    Compute a cross table between target and feature, and print 
    chi-suqared test results from this table.
    If the feature is quantitative, first it's segmented into bins.
    Parameters:
    - feature: pd.Series or array-like object.
        An object containing feature values
    - target: pd.Series or array-like object.
        An object containing target values.
    - categorical: bool.
        True if the feature is categorical, False otherwise.
        Default False.
    - nbins: int.
        Number of bins in which divide the feature. 
        If categorical is True, this parameter is ignored.
        Default 10.        
    """
    if(categorical):
        feature_discr = feature
    else:
        feature_discr = pd.cut(feature,nbins)
    cross_table = pd.crosstab(feature_discr, target)
    print(chi2_contingency(cross_table))
        
    
```

## Prediction model
A logistic regression model is used.
In order to evaluate performances, we perform hold-out technique (train test split).

```python
from sklearn.model_selection import train_test_split
X = data_clean.drop("Response",axis=1)
y = data_clean["Response"]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

### Dealing with imbalance
As seen from Response distriution, the dataset is imbalanced, where Negative class is much more represented than positive one. 
This could easly lead to overfitting. 

To avoid that, here an Oversampling thechinque is used. 
We prefer to generate new synthetic data, rather than duplicate observation, to avoid further overfitting (adding copies is overfitting prone, see [here](https://machinelearningmastery.com/tactics-to-combat-imbalanced-classes-in-your-machine-learning-dataset/)). 

We use a SMOTE variant, since in the dataset there are numeric and categorical features.

This variant is called SMOTENC, and it's implemented in [imbalanced-learn library](https://imbalanced-learn.org/stable/over_sampling.html).

Check [here](https://imbalanced-learn.org/stable/references/generated/imblearn.over_sampling.SMOTENC.html#imblearn.over_sampling.SMOTENC) to see documentation. 

Remember that oversampling must be applied only to train set.

```python
from imblearn.over_sampling import SMOTENC
smote_nc = SMOTENC(categorical_features=["Region_Code","Vehicle_Age",
                                         "Vehicle_Damage","Policy_Sales_Channel"])
X_train_smote,y_train_smote = smote_nc.fit_resample(X_train,
                                           y_train)

#type(data_smote)
```

Logistic regression fit algorithms require data standardization, but it must be applied only to quantitative variables (that is only Annual_Premium, now the column with index = 4) and to ordinal categorical variables, after label encoding (Vehicle_Age, now column wiuth index = 2).

Binary qualitative features require only label encoding to 0/1 values. 

Qualitative nominal features (Region_Code and Policy_Sales_Channel), require instead a OneHotEncoding.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_nominal_features = ["Region_Code", "Policy_Sales_Channel"]

#Vehicle_Age will be standardized since encoding reflects values sorting.
ordinal_and_continuous_features = ["Annual_Premium", "Vehicle_Age"]

ct = ColumnTransformer(
    [
        ("ohe", OneHotEncoder(sparse_output=False, handle_unknown="ignore"), categorical_nominal_features),
        ("ss", StandardScaler(), ordinal_and_continuous_features)
    ],
    #keep columns already transformed (Gender and Vehicle_Damage)
    remainder="passthrough" 
)

X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

### Model creation and fitting

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)

model.fit(X_train,y_train)
```

We want to optimize the classification threshold, so we don't use predict method to predict target (predict method uses a standard threshold of 0.5).

```python
#y_pred_train_original = model.predict(X_train)
y_pred_prob_train = model.predict_proba(X_train)

#y_pred_test_original = model.predict(X_test)
y_pred_prob_test = model.predict_proba(X_test)

model.classes_
```

Classification threshold optimization through ROC curve (computed on test set).

```python
from sklearn.metrics import roc_curve
#positive class scores are y_pred_prob_test[:,1] because positive class is
#in second position between model's classes 
#(you can check with model.classes_)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob_test[:,1], 
                                         pos_label=1,
                                 drop_intermediate=False)
plt.plot(fpr,tpr,'-b', label="ROC curve")
plt.plot(np.arange(0,1.1,0.1),np.arange(0,1.1,0.1),':k',label="baseline")
plt.grid()
plt.title("Test set ROC")
plt.xlabel("FPR")
plt.ylabel("TPR")

#this section is explained below the chart
#Euclidean distances between each ROC point and (0,1) point
distances = np.sqrt((1-tpr)**2 + fpr**2)

# find point with minimum distance from (0,1)
index = np.argmin(distances)
#and the corrsponding threshold
best_threshold = thresholds[index]

print(f"Best threshold: {best_threshold}")
plt.scatter(fpr[index],tpr[index],c="r",label="Best threshold")
plt.legend()
plt.show()
```

Roc curve is quite far from ideal case...

However, best threshold is detected with distance criterion:

one must find the point on ROC that is nearest to point (0,1), using euclidean distance as distance metric; after that, the threshold corresponding to this point is the best one. 

This threshold is displayed in above chart as a red dot.

### Model evaluation
Using the best threshold let's make predictions and evaluation with confusion matrix based scores.

```python
y_pred_train = np.where(y_pred_prob_train[:,1]>=best_threshold,1,0)
y_pred_test  = np.where(y_pred_prob_test[:,1]>=best_threshold,1,0)
```

```python
def display_confusion_matrix(cm, set_type="train"):
    """
    Plot a confusion matrix in seaborn heatmap style.
    Parameters:
      - cm : array-like object with dimension (n_classes,n_classes).
        The confusion matrix to plot.
      - set_type : str.
        Name of set type to plot (e.g. train or test). It's used for title.
        Default "train".    
    """
    
    plt.figure(figsize=[6.4, 4.8],dpi=100)
    sns.heatmap(
    cm,
    annot = True,
    square = True,
    cmap = "inferno",
    linewidth=.5,
    fmt = ".0f", #disable scientific notation,
    cbar = False
)

    plt.xlabel("Predicted values")
    plt.ylabel("True values")
    plt.title(f"Confusion matrix for {set_type} set")
    plt.show()
    # confusion matrix structure:
    # TP | FN
    # FP | TN
```

```python
from sklearn.metrics import confusion_matrix

# confusion matrix structure:
# TP | FN
# FP | TN

cm_train = confusion_matrix(y_train, y_pred_train, labels=[1,0])
cm_train = pd.DataFrame(cm_train,
                           index = [1,0],
                      columns = [1,0])

cm_test = confusion_matrix(y_test, y_pred_test, labels=[1,0])
cm_test = pd.DataFrame(cm_test,
                           index = [1,0],
                      columns = [1,0])

display_confusion_matrix(cm_train)
display_confusion_matrix(cm_test,"test")
```

Confusion matrix contains all info about predction, but it's hard to make some consideration. For this reason it's more useful to compute some metrics from it.  

Metrics used here are:
- Accuracy
- Sensitivity (recall)
- Specificity
- Precision
- F1-score


```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def specificity_score(cm):
    """
    Compute the specificity score for a 2x2 confusion matrix.
    Parameters:
      - cm : ndarray of shape (2,2).
        The binary confusion matrix.
        It must have this structure:
         TP | FN
         FP | TN
    
    return: 
      - specificity = TN / (TN + FP) as float.
    """
    
    TP, FN, FP, TN = cm.ravel()
    specificity_value = TN / (TN + FP)
    return specificity_value

def classification_metrics(y_true,y_pred):
    """
    Print some metrics of a binary classification.
    Metrics printed are:
    Accuracy
    Sensitivity (recall)
    Specificity
    Precision
    F1 score
    Parameters:
      - y_true : 1d array-like object.
        Array with correct labels.
      - y_pred: 1d array-like object.
        Array with predicted labels.
    """
    acc = accuracy_score(y_true,y_pred)
    prec = precision_score(y_true,y_pred,pos_label=1)
    sensitivity = recall_score(y_true,y_pred,pos_label=1)
        
    cm = confusion_matrix(y_true,y_pred, labels=[1,0])
    specificity = specificity_score(cm)
    f1 = f1_score(y_true,y_pred)
        
    print(f"Accuracy: {acc:.3f}")
    print(f"Sensitivity: {sensitivity:.3f}")
    print(f"Specificity: {specificity:.3f}")
    print(f"Precision: {prec:.3f}")
    print(f"F1 score: {f1:.3f}")
```

```python
print("Train set")
classification_metrics(y_train,y_pred_train)
print("-"*20)
print("Test set")
classification_metrics(y_test,y_pred_test)
```

Sensitivity and Specificity in test test and train set are very similar, and it's a good sign about generalization: smote avoids model overfitting.
However note as precision is quite low in test test.

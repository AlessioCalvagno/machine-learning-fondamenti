### 2. crea modello di regressione lineare semplice con feature più correlata col target (che è price)

tale variabile è LSTAT che ha correlazione del 74% (in modulo).

```python
#modello di regressione lineare semplice
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(df["LSTAT"].values.reshape(-1,1),df["PRICE"].values)
```

Valutazione del modello lineare appena ottenuto (uso le funzioni già
definite in sklearn)

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

x_train = df["LSTAT"].values.reshape(-1,1)
y_train = df["PRICE"].values
y_pred = lr.predict(x_train)

mse = mean_squared_error(y_train,y_pred)
rmse = mse**0.5
mae = mean_absolute_error(y_train,y_pred)
r2 = r2_score(y_train,y_pred)

print(f"Performances:\nMSE: {mse:.2f}\nRMSE: {rmse:.2f}\nMAE: {mae:.2f}\nR2: {r2:.2f}")
```

Il modello ottenuto è così così, con $R^2 = 54\%$ 

### 3. crea modello di regressione lineare multipla con le due variabili più correlate col target (price)

Come prima, bisogna gestire i valori mancanti e il reshape.

Le due variabili èpiù correlate sono LSTAT (già gestita) e RM.

```python
x_train_multi = df[["LSTAT","RM"]].values #reshape non necessario?

lr_multi = LinearRegression()
lr_multi.fit(x_train_multi,y_train)

#metriche
y_pred_multi = lr_multi.predict(x_train_multi)

mse_multi = mean_squared_error(y_train,y_pred_multi)
rmse_multi = mse_multi**0.5
mae_multi = mean_absolute_error(y_train,y_pred_multi)
r2_multi = r2_score(y_train,y_pred_multi)

print(f"Performances:\nMSE: {mse_multi:.2f}\nRMSE: {rmse_multi:.2f}\nMAE: {mae_multi:.2f}\nR2: {r2_multi:.2f}")
```

Le performances migliorano, le funzioni di costo diminuiscono e ho $R^2=64\%$

### 4. Crea modello multiplo con 3 variabili (segui sempre la stessa procedura), ma qui introduci anche termini polinomiali. Prova tutti i gradi da 1 a 5.

Passo ad una relazione non lineare (polinomiale), provando i possibili gradi da 1 a 5. Si gestisce tutto con il transformer PolynomialFeatures di sklearn.

Le 3 variabili più correlate sono LSTAT, RM, e PTRATIO.

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

degrees = np.arange(1,6) #da 1 a 5

x_train = df[["LSTAT","RM","PTRATIO"]].values #array numpy

print("Polynomial regression with bias")

for degree in degrees:
    pf = PolynomialFeatures(degree) 
    #PolynomialFeatures(degree,include_bias=False)
    x_p = pf.fit_transform(x_train)
    lr_d = LinearRegression()
    lr_d.fit(x_p,y_train)
    
    y_pred_d = lr_d.predict(x_p)
    
    #metriche
    mse_d = mean_squared_error(y_train,y_pred_d)
    rmse_d = mse_d**0.5
    mae_d = mean_absolute_error(y_train,y_pred_d)
    r2_d = r2_score(y_train,y_pred_d)
    
    print(f"Metrics for degree {degree}:\nMSE: {mse_d:.2f}\nRMSE: {rmse_d:.2f}\nMAE: {mae_d:.2f}\nR2: {r2_d:.2f}")
    print("-"*10)
    
    
    
```

```python
print("Polynomial regression without bias")

for degree in degrees:
    #pf = PolynomialFeatures(degree) 
    pf =PolynomialFeatures(degree,include_bias=False)
    x_p = pf.fit_transform(x_train)
    lr_d = LinearRegression()
    lr_d.fit(x_p,y_train)
    
    y_pred_d = lr_d.predict(x_p)
    
    #metriche
    mse_d = mean_squared_error(y_train,y_pred_d)
    rmse_d = mse_d**0.5
    mae_d = mean_absolute_error(y_train,y_pred_d)
    r2_d = r2_score(y_train,y_pred_d)
    
    print(f"Metrics for degree {degree}:\nMSE: {mse_d:.2f}\nRMSE: {rmse_d:.2f}\nMAE: {mae_d:.2f}\nR2: {r2_d:.2f}")
    print("-"*10)
```
